## Copying your Colab Project to Git

To move your project to a Git repository, you'll generally follow these steps:

1.  **Configure Git:** Set your user name and email.
2.  **Initialize a local Git repository:** This creates a `.git` directory to track changes.
3.  **Add your files:** Stage the changes you want to commit.
4.  **Commit your changes:** Save a snapshot of your project's current state.
5.  **Add a remote repository:** Link your local repository to a remote one (e.g., on GitHub, GitLab, or Bitbucket).
6.  **Push your changes:** Upload your committed changes to the remote repository.

### Important Considerations:

*   **Sensitive Information:** Be careful not to upload sensitive data like API keys, passwords, or client secrets to public repositories. You should use `.gitignore` to exclude such files or leverage Colab's secret management.
*   **Existing `.ipynb`:** Colab notebooks are `.ipynb` files. Git tracks these just like any other file.
*   **GitHub/GitLab:** Before pushing, you'll need to create an empty repository on your chosen Git hosting service (e.g., GitHub, GitLab) and get its URL.

```python
import requests
import base64
import json
import uuid

# Замените эти значения своими данными
client_id = '019e45bc-d921-7dda-94ea-01c8df810272'
client_secret = 'dc4d19c5-865d-42fe-8a88-7700e87e94e0'
scope = 'GIGACHAT_API_PERS' # или другой нужный вам scope

req_id = str(uuid.uuid4())

# Создаем Authorization Key
auth_key = base64.b64encode(f"{client_id}:{client_secret}".encode()).decode()

# Данные для запроса токена
payload_token = {'scope': scope}
headers_token = {
    'Content-Type': 'application/x-www-form-urlencoded',
    'Accept': 'application/json',
    'Authorization': f'Basic {auth_key}',
    'RqUID': str(uuid.uuid4()) # Уникальный идентификатор запроса
}

# Отправляем запрос на получение токена
token_url = 'https://ngw.devices.sberbank.ru:9443/api/v2/oauth'
token_response = requests.post(token_url, headers=headers_token, data=payload_token, verify=False)

access_token = None
if token_response.status_code == 200:
    access_token_data = token_response.json()
    access_token = access_token_data['access_token']
    print(f"Ваш токен доступа: {access_token}")
else:
    print(f"Ошибка получения токена: {token_response.text}")

# --- Используем полученный токен для отправки запроса к GigaChat API ---

if access_token:
    gigachat_api_url = "https://gigachat.devices.sberbank.ru/api/v1/chat/completions"

    # Пример payload для запроса к GigaChat API
    gigachat_payload = json.dumps({
        "model": "GigaChat-2-Max", # или другая доступная модель
        "messages": [
            {
                "role": "user",
                "content": "Привет! Расскажи мне что-нибудь интересное."
            }
        ],
        "profanity_check": True
    })

    gigachat_headers = {
        'Content-Type': 'application/json',
        'Accept': 'application/json',
        'Authorization': f'Bearer {access_token}'
    }

    # Отправляем запрос к GigaChat API
    try:
        gigachat_response = requests.post(gigachat_api_url, headers=gigachat_headers, data=gigachat_payload, verify=False)
        gigachat_response.raise_for_status() # Вызывает исключение для HTTP ошибок (4xx или 5xx)
        print("\nОтвет от GigaChat:")
        print(json.dumps(gigachat_response.json(), indent=2, ensure_ascii=False))
    except requests.exceptions.RequestException as e:
        print(f"\nОшибка при запросе к GigaChat API: {e}")
        if hasattr(e, 'response') and e.response is not None:
            print(f"GigaChat API Response content: {e.response.text}")
else:
    print("Не удалось получить access_token, запрос к GigaChat API не будет выполнен.")
```
